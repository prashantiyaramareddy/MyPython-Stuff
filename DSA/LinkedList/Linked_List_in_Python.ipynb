### Search in Linked List

```python
class Node:
  def __init__(self, k):
    self.key = k
    self.next = None

  def search(head, x):
    pos = 1
    curr = head
    while curr != None:
      if curr.key == x:
        return pos
      pos = pos + 1
      curr = curr.next
    return -1


head = Node(10)
head.next = Node(20)
head.next.next = Node(15)
head.next.next.next = Node(30)
print(Node.search(head, 15))
```

### Time Complexity  and Auxillary Space of Search in Linkedlist

Time Complexity - O(n)

Auxillary Space - O(1)

### Inserting at the end of the linked list

```python
class Node:
  def __init__(self, key):
    self.key = key
    self.next = None

  def insertEnd(head, key):
    new_node = Node(key)
    if head is None:
      return new_node

    curr = head
    while curr.next != None:
      curr = curr.next
    curr.next = new_node
    return head


  def printList(head):
    curr = head
    while curr!= None:
        print(curr.key)
        curr = curr.next

# Driver code
head = None
head = Node.insertEnd(head, 10)
head = Node.insertEnd(head, 20)
head = Node.insertEnd(head, 30)
head = Node.insertEnd(head, 40)
Node.printList(head)
```

Complexity Analysis:

* Time complexity: O(N), where N is the number of nodes in the linked list. Since there is a loop from head to end, the function does O(n) work.
This method can also be optimized to work in O(1) by keeping an extra pointer to the tail of the linked list/

* Auxiliary Space: O(1)

### Delete Last Node of the Linked list

```python
class Node:
  def __init__(self, key):
    self.key = key
    self.next = None

  def deleteLast(head):
    if head == None:
      return None
    if head.next == None:
      return None
    curr = head
    while curr.next.next != None:
      curr = curr.next
    curr.next = None
    return head

  def printList(head):
    curr = head
    while curr!= None:
        print(curr.key)
        curr = curr.next
    print()

# Driver code
head = Node(10)
head.next = Node(20)
head.next.next = Node(30)
head.next.next.next = Node(40)
head.next.next.next.next = Node(50)

Node.printList(head)

head1 = Node.deleteLast(head)

Node.printList(head1)
```

### Sorted insert in a linked list

```python
class Node:
  def __init__(self, key):
    self.key = key
    self.next = None

  def sortedInsert(head,x):
    temp = Node(x)
    if head == None:
      return temp
    if x <= head.key:
      temp.next = head
      return temp
    curr = head
    while curr.next != None and curr.next.key < x:
      curr = curr.next
    # Insert temp after curr
    temp.next = curr.next
    curr.next = temp
    return head


  def printList(head):
    curr = head
    while curr!= None:
        print(curr.key)
        curr = curr.next
    print()

# Driver code
head = Node(10)
head.next = Node(20)
head.next.next = Node(30)
head.next.next.next = Node(40)
head.next.next.next.next = Node(50)

Node.printList(head)

head = Node.sortedInsert(head,39)

Node.printList(head)
```

### Middle of a linked list - Naive and efficient approach

```python
class Node:
  def __init__(self, key):
    self.key = key
    self.next = None

  def printMiddleNaive(head):
    if head == None:
      return
    count = 0
    curr = head
    while curr != None:
      count = count + 1
      curr = curr.next
    curr = head
    for i in range(count//2):
      curr = curr.next
    print(curr.key)
    return

  def printMiddle(head):
    if head == None:
      return
    slow = head
    fast = head
    while fast != None and fast.next != None:
      slow = slow.next
      fast = fast.next.next
    print(slow.key)

    return

  def printList(head):
    curr = head
    while curr!= None:
        print(curr.key)
        curr = curr.next
    print()

# Driver code
head = Node(10)
head.next = Node(20)
head.next.next = Node(30)
head.next.next.next = Node(40)
head.next.next.next.next = Node(50)

Node.printList(head)

Node.printMiddle(head)
```

### Nth Node from end of linked list

```python
class Node:
  def __init__(self, key):
    self.key = key
    self.next = None

  def printNNaive(head,n):
    count = 0
    curr = head
    while curr:
      curr = curr.next
      count = count + 1
    if count < n:
      return
    curr = head
    for i in range(1,count-n+1):
      curr = curr.next
    print(curr.key)

  def printNEfficient(head,n):
    if head == None:
      return
    first = head
    second = head
    for i in range(n):
      if first == None:
        return
      first = first.next

    while first != None:
      first = first.next
      second = second.next

    print(second.key)

  def printList(head):
    curr = head
    while curr!= None:
        print(curr.key)
        curr = curr.next
    print()

# Driver code
head = Node(10)
head.next = Node(20)
head.next.next = Node(30)
head.next.next.next = Node(40)
head.next.next.next.next = Node(50)

Node.printList(head)

Node.printNNaive(head,3)

Node.printNEfficient(head,5)
```

### Reverse a linked list in Python

## Using Stack Method

```python
class Node:
    def __init__(self, k):
        self.key = k
        self.next = None


def reverseList(head):
    stack = []

    curr = head

    while curr is not None:
        stack.append(curr.key)
        curr = curr.next

    curr = head
    while curr is not None:
        curr.key = stack.pop()
        curr = curr.next

    return head


def printList(head):
    curr = head
    while curr != None:
        print(curr.key)
        curr = curr.next
    print()


head = Node(10)
head.next = Node(20)
head.next.next = Node(30)
head.next.next.next = Node(40)

printList(head)

head = reverseList(head)

printList(head)
```

## Efficient Method

```python
class Node:
    def __init__(self, k):
        self.key = k
        self.next = None


def reverseList(head):
    curr = head
    prev = None

    while curr is not None:
        next = curr.next
        curr.next = prev
        prev = curr
        curr = next

    return prev


def printList(head):
    curr = head
    while curr != None:
        print(curr.key)
        curr = curr.next
    print()


head = Node(10)
head.next = Node(20)
head.next.next = Node(30)
head.next.next.next = Node(40)

printList(head)

head = reverseList(head)

printList(head)
```

## 1. Stack Method (or Two-Pass Method)

This method typically involves two passes over the list. The stack is generally an analogy for any method that requires storing all nodes (like a list or array) or requires two full list traversals (one for counting, one for traversing to the middle).

### â³ Time Complexity: $O(N)$

* **First Pass:** You traverse the entire list to count the total number of nodes, $N$. This takes $O(N)$ time.
* **Second Pass:** You traverse the list again up to the middle position ($N/2$). This takes $O(N/2)$, which is still $O(N)$ time.
* **Total Time:** $O(N) + O(N) = O(N)$.

### ðŸ’¾ Space Complexity: $O(N)$

* If you use an **actual stack** or array to store all nodes during the first pass (which is a common implementation when comparing to the efficient method), you need space proportional to the number of nodes, $N$.
* If you just count the nodes and then do a second pass, the space complexity is $O(1)$ (constant extra space). However, if the "stack method" you are referring to involves collecting all nodes, the complexity is $O(N)$.
